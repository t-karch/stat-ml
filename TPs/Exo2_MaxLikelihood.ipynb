
## Exercice 2 : maximum de vraisemblance

### <span style="color:blue">QUESTION 2.1 :</span> 

Tirer 10 fois une pièce à pile ou face et modéliser les résultats obtenus comme ceux d'une variable aléatoire X qui vaut X_i=0 si on a pile et X_i=1 si on a face.

Calculez le maximum de vraisemblance du paramètre p d'un loi de Bernoulli qui modéliserait le problème. Pour y arriver, différentes valeures possibles de p seront testées et le p retenu sera celui qui a la plus grande vraisemblance.


- Vérifier empiriquement comment évolue ce maximum de vraisemblance si l'on effectue de plus en plus de tirages
- Que se passe-t-il quand il y a trop de tirages ? Représenter la log-vraisemblance plutot que la vraisemblance dans ce cas.




```python
NbTirages=100
NbPiles = np.random.randn(NbTirages) > 0
NbPiles = np.sum(NbPiles)
NbFaces=NbTirages-NbPiles


PossibleValuesForP=np.linspace(0.01,0.99,100)


def vraisemblance(n_pile, n_face, p):
    """
    n_pile : int
        Nombre de piles observés
    n_face : int
        Nombre de faces observés
    p : float   
    """


    return (p ** n_pile) * ((1 - p) ** n_face)

max_vraisemblance=-10e10
argmax_vraisemblance=0

log_vraisemblance_bool=False


if not log_vraisemblance_bool:
    for p in PossibleValuesForP:
        print(vraisemblance(NbPiles,NbFaces,p) )
        if vraisemblance(NbPiles,NbFaces,p) > max_vraisemblance:
            max_vraisemblance=vraisemblance(NbPiles,NbFaces,p)
            argmax_vraisemblance=p

    print("la valeur la plus vraisemblable est",argmax_vraisemblance)

# si log_vraisemblance_bool est vrai, on utilise la log-vraisemblance
if log_vraisemblance_bool:
    def log_vraisemblance(n_pile, n_face, p):
        
        return n_pile * np.log(p) + n_face * np.log(1 - p)


    for p in PossibleValuesForP:
        print(log_vraisemblance(NbPiles, NbFaces, p) )
        if log_vraisemblance(NbPiles, NbFaces, p) > max_vraisemblance:
            max_vraisemblance=log_vraisemblance(NbPiles, NbFaces, p)
            argmax_vraisemblance=p
    print("la valeur la plus vraisemblable est p=",argmax_vraisemblance)



```

Plus on augmente le nombre de tirages, plus la vraissemblance diminue (jusqu'à arriver a un "0.0" informatique pour X> 1000 par exemple).

On trouve que le maximum de vraisemblance est pour un p extrème (soit 0.01 ou 0.99 en fnction des tests).
