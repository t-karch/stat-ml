### Question 1.2 

Afin de répondre aux questions, deux hypothèses sont formulées :
- Pour chaque marque de moteur, la relation entre la concentration et le rendement est supposée linéaire.
- La distribution du bruit des observations est supposé suivre une loi Normale centrée.

Quelle démarche de modélisation statistique allez-vous utiliser pour répondre aux trois questions posées en introduction ? Une fois la démarche bien expliquée, testez la sur les données observées afin de conclure sur l'efficacité du produit.

Vu que la relation entre la concentration et le rendement est supposée linéaire et que l'erreur suit une loi Normal centrée, on peut modéliser notre régression linéaire à l'aide de LinearRegression().

Pour voir une difference entre les deux moteurs on entraine deux regressions linéaires, une pour chaque marque. 

Pour la question 1 on vérifiera que les pentes sont non nulles et on fera un test d'hypothèse pour vérifier qu'elles le sont significativement. Pour vérifier cela on effectue le test d'hypothèse avec α = 0.05:\
   **H0**: La pente = 0 (le produit n'a pas d'effet)\
   **H1**: La pente ≠ 0 (le produit a un effet)\
et ce pour les deux pentes.


Pour la question 2, pour voir s'il y a une difference on comparera les pentes, et de meme pour voir si cette difference est significative on fera le test d'hypothèse suivant :\
**H0**: les pentes sont égales\
**H1**: les pentes ne sont pas égales

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy import stats

# régression linéaire pour chaque marque
lr_brand1 = LinearRegression()
lr_brand1.fit(df_brand1[['Concentration']], df_brand1['Efficiency'])
efficiency_pred_b1 = lr_brand1.predict(df_brand1[['Concentration']])
a_b1 = lr_brand1.coef_[0]
b_b1 = lr_brand1.intercept_

lr_brand2 = LinearRegression()
lr_brand2.fit(df_brand2[['Concentration']], df_brand2['Efficiency'])
efficiency_pred_b2 = lr_brand2.predict(df_brand2[['Concentration']])
a_b2 = lr_brand2.coef_[0]
b_b2 = lr_brand2.intercept_

# visualisation des résultats
plt.scatter(df_brand1[['Concentration']], df_brand1['Efficiency'], color='blue', label='Brand 1 Data')
plt.plot(
    df_brand1['Concentration'],
    efficiency_pred_b1,
    color='blue', label='Brand 1 Fit'
)
plt.scatter(df_brand2[['Concentration']], df_brand2['Efficiency'], color='red', label='Brand 2 Data')
plt.plot(
    df_brand2['Concentration'],
    efficiency_pred_b2,
    color='red', label='Brand 2 Fit'
)
plt.xlabel('Concentration')
plt.ylabel('Efficiency')
plt.title('Efficiency vs Concentration')
plt.legend()
plt.show()

# calcul des MSE pour chaque modèle
mse_brand1 = mean_squared_error(df_brand1[['Efficiency']], efficiency_pred_b1)
mse_brand2 = mean_squared_error(df_brand2[['Efficiency']], efficiency_pred_b2)

print(f"Le MSE pour Brand 1: {mse_brand1}\nLe MSE pour Brand 2: {mse_brand2}")

print(f"Les pentes sont respectivement {a_b1:.4f} pour Brand 1 et {a_b2:.4f} pour Brand 2")



# Q2 calcul de la statistique t pour tester l'hypothèse nulle que les pentes sont égales
# H0: a_b1 = a_b2
# H1: a_b1 != a_b2
n1 = len(df_brand1)
n2 = len(df_brand2)

# Calculs corrects des erreurs standard pour les pentes
def calculate_slope_se(X, y, mse):
    """Calcule l'erreur standard de la pente"""
    n = len(X)
    X_centered = X - np.mean(X)
    sum_x_squared = np.sum(X_centered**2)
    se_slope = np.sqrt(mse / sum_x_squared)
    return se_slope

# Calcul correct des erreurs standard
se1 = calculate_slope_se(df_brand1['Concentration'].values, df_brand1['Efficiency'].values, mse_brand1)
se2 = calculate_slope_se(df_brand2['Concentration'].values, df_brand2['Efficiency'].values, mse_brand2)


se_diff = np.sqrt(se1**2 + se2**2)
t_stat = (a_b1 - a_b2) / se_diff

dl = n1 + n2 - 4
p_value_ttest = 2 * (1 - stats.t.cdf(np.abs(t_stat), dl))   
print(f"t-statistic: {t_stat:.4f}, p-value: {p_value_ttest:.4f}")
```

```python
# Test d'hypothèse pour Q1 : Les pentes sont-elles significativement différentes de zéro ?

print("--- TEST POUR BRAND 1 ---")
t_stat_b1 = (a_b1 -0 ) / se1 
df_b1 = n1 - 2 
p_value_b1 = 2 * (1 - stats.t.cdf(np.abs(t_stat_b1), df_b1))

print(f"Pente observée: {a_b1:.6f}")
print(f"Erreur standard: {se1:.6f}")
print(f"Statistique t: {t_stat_b1:.4f}")
print(f"Degrés de liberté: {df_b1}")
print(f"P-value: {p_value_b1:.12f}")

if p_value_b1 < 0.05:
    print("REJET de H0 : La pente est significativement différente de zéro donc le produit a un effet significatif pour Brand 1")
else:
    print("ACCEPTATION de H0 : Pas de preuve d'effet significatif")
print()

# Test pour Brand 2
print("--- TEST POUR BRAND 2 ---")
t_stat_b2 = a_b2 / se2  # t = (pente - 0) / erreur_standard
df_b2 = n2 - 2  # degrés de liberté pour une régression simple
p_value_b2 = 2 * (1 - stats.t.cdf(np.abs(t_stat_b2), df_b2))

print(f"Pente observée: {a_b2:.6f}")
print(f"Erreur standard: {se2:.6f}")
print(f"Statistique t: {t_stat_b2:.4f}")
print(f"Degrés de liberté: {df_b2}")
print(f"P-value: {p_value_b2:.12f}")

if p_value_b2 < 0.05:
    print("REJET de H0 : La pente est significativement différente de zéro donc le produit a un effet significatif pour Brand 2")
else:
    print("ACCEPTATION de H0 : Pas de preuve d'effet significatif")
```

```python

# Question 2: Comparaison des pentes (H0: pente1 = pente2)
print("=== QUESTION 2: L'effet diffère-t-il entre les marques? ===")
print("Test d'hypothèse: H0: pente1 = pente2 vs H1: pente1 ≠ pente2")
print()

se_diff = np.sqrt(se1**2 + se2**2)
t_stat_diff = (a_b1 - a_b2) / se_diff
df_diff = n1 + n2 - 4
p_value_diff = 2 * (1 - stats.t.cdf(np.abs(t_stat_diff), df_diff))

print(f"Différence des pentes: {a_b1 - a_b2:.6f}")
print(f"Erreur standard différence: {se_diff:.6f}")
print(f"t-statistique: {t_stat_diff:.4f}")
print(f"p-value: {p_value_diff:.6f}")

if p_value_diff < 0.05:
    print("On rejette H0: Les pentes sont significativement différentes (p < 0.05)")
else:
    print("Pas de différence significative entre les pentes (p ≥ 0.05)")
print()

```

**Question 1 : "Le produit semble-t-il avoir un effet ?"**
- **Analyse visuelle** : Les pentes sont non nulles, l'efficacité augmente avec la concentration
- **Test statistique** : Test t sur chaque pente (H₀: pente = 0 vs H₁: pente ≠ 0)
- **Conclusion** : Le produit a un effet significatif sur l'efficacité des moteurs (les deux pentes sont significativement différentes de zéro avec p < 0.05)

**Question 2 : "Cet effet est-il différent en fonction de la marque du moteur ?"**  
- **Analyse visuelle** : Les pentes semblent différentes numériquement
- **Test statistique** : Test t de comparaison des pentes (H₀: pente₁ = pente₂ vs H₁: pente₁ ≠ pente₂)
- **Conclusion** : Les pentes sont siginficativement differentes

**Question 3 : "Cet effet dépend-il de la concentration du produit ?"**
- **Analyse** : Modèle de régression linéaire (relation concentration ↔ efficacité)
- **Conclusion** : Oui, l'efficacité augmente linéairement avec la concentration pour les deux marques. La relation est positive et significative.

#### QUESTION 2.2.2

Peut-on statistiquement affirmer qu'il existe une relation significative entre le niveau de stress et (indépendament) *sensor_01*, *sensor_12* ou bien *sensor_15* ? Si oui, décrivez votre procédure de test.


```python
sensors = [0, 11, 14] 
sensor_names = ["sensor_01", "sensor_12", "sensor_15"]

for idx, name in zip(sensors, sensor_names):
    X_sensor = X.iloc[:, idx].values.reshape(-1, 1)
    y = Y.values.ravel()
    lr = LinearRegression()
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    mse_scores = []
    for train_idx, test_idx in kf.split(X_sensor):
        lr.fit(X_sensor[train_idx], y[train_idx])
        y_pred = lr.predict(X_sensor[test_idx])
        mse_scores.append(mean_squared_error(y[test_idx], y_pred))
    print(f"{name}: MSE moyen (5-fold CV) = {np.mean(mse_scores):.3f}")

    # Affichage du nuage de points et de la droite de régression
    plt.scatter(X_sensor, y, label='Données')
    plt.plot(X_sensor, lr.predict(X_sensor), color='red', label='Régression')
    plt.xlabel(name)
    plt.ylabel('Stress')
    plt.title(f"Régression linéaire: {name} vs Stress")
    plt.legend()
    plt.show()
```

On a fait une validation croisé à l'aide d'un KFlod de 5 splits.
Le KFold semble être un bon choix de validation croisé car on vérifie que le modèle reste rosbuste même sur un plus petit jeu de données.

```python
from scipy.stats import pearsonr

sensors = [0, 11, 14]
sensor_names = ["sensor_01", "sensor_12", "sensor_15"]

for sensor_id, name in zip(sensors, sensor_names):
    x = X.iloc[:, sensor_id].values
    y = Y.values.ravel()
    corr, pval = pearsonr(x, y)
    print(f"{name}: corrélation = {corr:.3f}, p-value = {pval:.3e}")
    if pval < 0.05:
        print(f"Relation significative entre {name} et le stress (p < 0.05)")
    else:
        print(f"Aucune relation significative entre {name} et le stress (p >= 0.05)")
```

```python
# Q 2.3.1
x = X.iloc[:, 11].values
y = Y.values.ravel()


def s2(y_true, y_pred):
    n = len(y_true)
    SSE = np.sum((y_true - y_pred)**2)
    return SSE / (n-1)

def cook(x, y, i):
    lr = LinearRegression()
    xx = np.delete(x, i)
    yy = np.delete(y,i)
    lr.fit(xx[:, np.newaxis], yy)
    
    lr_tot = LinearRegression()
    lr_tot.fit(x[:, np.newaxis], y)
    y_tot = lr_tot.predict(x[:, np.newaxis])
    
    y_pred = lr.predict(x[:, np.newaxis])
    SSD = np.sum((y_pred - y_tot)**2)
    s_2 = s2(y_tot, y)
    
    return SSD / (2*s_2)

# affiche les distance de cook
for j in range(len(x)):
    Dj = cook(x, y, j)
    if Dj > 1:
        print(x[j],y[j])
        print(f" pour j={j},   Di={cook(x, y, j):.3f}")
```

##### Réponse Question 2.3.1
Pour detecter automatiquement les valeurs abérantes, nous avons calculer les distances de cook et nous les avons comparé a 1. Si jamais la distance de cook est supérieur à cette valeur, alors on considère le point comme une donnée abérante.

Nous souhaitons maintenant sélectionner automatiquement un nombre réduit, mais supérieur à 1, de capteurs qui nous permettraient de prédire au mieux le niveau de stress. Nous allons pour cela utiliser la régression multiple avec un terme de régularisation.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error

# Suppression des deux outliers détectés (indices à adapter selon votre détection)
outlier_indices = [2, 62]
X_clean = X.drop(index=outlier_indices).reset_index(drop=True)
Y_clean = Y.drop(index=outlier_indices).reset_index(drop=True)

# Centrage-réduction des capteurs
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)

from sklearn.model_selection import KFold

#-> 1ere passe : estimtion de l'echelle du alpha optimal

NB_folds=8

for alpha in [0.001,0.01,0.1,1.,10.]:
  sum_mse_scores=0.
  kf = KFold(n_splits=NB_folds)
  for train, test in kf.split(X_clean):
    X_train = X_scaled[train, :]
    Y_train = Y_clean.values[train]
    X_test = X_scaled[test, :]
    Y_test = Y_clean.values[test]

    lasso_regressor = Lasso(alpha=alpha)
    lasso_regressor.fit(X_train, Y_train)

    Y_pred_lasso = lasso_regressor.predict(X_test)
    mse_score_lasso = mean_squared_error(Y_test, Y_pred_lasso)
    sum_mse_scores+=mse_score_lasso
    

  print(alpha," mean MSE: ",sum_mse_scores/NB_folds)

# on observe que alpha = 0.1 nous donne le plus petit MSE
# on va faire une deuxieme passe autour de 0.1

```

```python
# deuxieme passe
for alpha in [0.025,0.05,0.1,0.2,0.4,0.5,0.6,0.8]:
  sum_mse_scores=0.
  kf = KFold(n_splits=NB_folds)
  for train, test in kf.split(X_clean):
    X_train = X_scaled[train, :]
    Y_train = Y_clean.values[train]
    X_test = X_scaled[test, :]
    Y_test = Y_clean.values[test]

    lasso_regressor = Lasso(alpha=alpha)
    lasso_regressor.fit(X_train, Y_train)

    Y_pred_lasso = lasso_regressor.predict(X_test)
    mse_score_lasso = mean_squared_error(Y_test, Y_pred_lasso)
    sum_mse_scores+=mse_score_lasso
    

  print(alpha," mean MSE: ",sum_mse_scores/NB_folds)

# Notre meilleur alpha est donc 0.4

```

```python
# Entrainement avec alpha optimal
lasso_regressor = Lasso(alpha=0.4)
lasso_regressor.fit(X_scaled, Y_clean)
plt.plot(lasso_regressor.coef_, 'r.')
plt.xlabel("Capteur")
plt.ylabel("Coefficient Lasso")
plt.title("Coefficients des capteurs sélectionnés (Lasso Régularisé)")
plt.show()

# Selection des capteurs ayant des coefficients non nuls
selected = np.where(lasso_regressor.coef_ != 0)[0]
print("Capteurs sélectionnés (indices):", selected)
print("Coefficients Lasso:", lasso_regressor.coef_[selected])
```
